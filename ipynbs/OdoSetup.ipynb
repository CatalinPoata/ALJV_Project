```python
!mkdir ./pytorch-CycleGAN-and-pix2pix/datasets/trainA
!mkdir ./pytorch-CycleGAN-and-pix2pix/datasets/trainB
!mkdir ./pytorch-CycleGAN-and-pix2pix/datasets/testA
!mkdir ./pytorch-CycleGAN-and-pix2pix/datasets/testB

from sklearn.model_selection import train_test_split
import os
import shutil
import pandas as pd

data_A = "./trainA"
data_B = "./trainB"

df_A = pd.DataFrame(columns=["filepath"])
df_B = pd.DataFrame(columns=["filepath"])

for file_A in os.listdir(data_A):
  df_A.loc[len(df_A)] = [file_A]

for file_B in os.listdir(data_B):
  df_B.loc[len(df_B)] = [file_B]

df_A_train, df_A_test = train_test_split(df_A, test_size=0.2, random_state=42)
df_B_train, df_B_test = train_test_split(df_B, test_size=0.2, random_state=42)

for index, row in df_A_train.iterrows():
  shutil.copy(os.path.join(data_A, row["filepath"]), os.path.join("./pytorch-CycleGAN-and-pix2pix/datasets/trainA", row["filepath"]))

for index, row in df_B_train.iterrows():
  shutil.copy(os.path.join(data_B, row["filepath"]), os.path.join("./pytorch-CycleGAN-and-pix2pix/datasets/trainB", row["filepath"]))

for index, row in df_A_test.iterrows():
  shutil.copy(os.path.join(data_A, row["filepath"]), os.path.join("./pytorch-CycleGAN-and-pix2pix/datasets/testA", row["filepath"]))

for index, row in df_B_test.iterrows():
  shutil.copy(os.path.join(data_B, row["filepath"]), os.path.join("./pytorch-CycleGAN-and-pix2pix/datasets/testB", row["filepath"]))
```
